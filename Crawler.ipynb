```python
import json, time
import requests
from time import sleep
import random

links = []
allData = []
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'
}

# 生成API連結
# movie_ids: http://files.tmdb.org/p/exports/movie_ids_08_08_2021.json.gz
def getid():
    # 開啟id檔
    with open('../movie_ids.json','r',encoding="utf-8") as fp:
        data = json.loads("[" + fp.read().replace("}\n{", "},\n{") + "]")
        for itm in data :
            film_id = itm['id']
            link = f'https://api.themoviedb.org/3/movie/{film_id}?api_key=2711e2a51f9e9d7e153fd5d6c4f4263d&append_to_response=credits'
            links.append({'title': itm['original_title'], 'link': link})
    with open("filename.json", "w", encoding="utf-8") as f:
        f.write(json.dumps(links, ensure_ascii=False))

def writeData():
    with open('filename.json', encoding='utf-8') as f:
        listResult = json.loads(f.read())
    
    for i in listResult:
        r = requests.get(i['link'], headers = headers)
        obj = json.loads(r.text)
        allData.append(obj)
        print(i['title'])     
    with open('filedata.json', 'w', encoding='utf-8') as f:
        f.write(json.dumps(allData, ensure_ascii=False) )


if __name__ == "__main__":
    time1 = time.time()
    getid()
    writeData()
    print(f"執行總花費時間: {time.time() - time1}")
```
